```python
####  run on python local
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False

x= df1['日期']

y1=df1['全區']
y2=df1['萬華區']
y3=df1['文山區']
y4=df1['大同區']
y5=df1['士林區']

fig = plt.figure()
plt.plot(x,y1,label="全區")
plt.plot(x,y2,label="萬華區")
plt.plot(x,y3,label="文山區")
plt.plot(x,y4,label="大同區")
plt.plot(x,y5,label="士林區")

plt.title("410701012",fontsize=20)
plt.xlabel("日期",fontsize=20)
plt.ylabel("確診人數",fontsize=20)
plt.ylim([0, 320])
plt.legend()
fig.autofmt_xdate()
plt.show()

```
